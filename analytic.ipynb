```python
from aad import new_leaf
import numpy as np
import pandas as pd
from main import my_plot, bs_analytic
```

```python
spots = np.linspace(40.00, 220.00, 32)
r = 0.02
T = 1.00
v0 = 0.40
K = 100.00
values: list[list[float]] = []

for fspot in spots:
    spot = new_leaf(fspot)
    strike = new_leaf(K)
    vol = new_leaf(v0)
    rate = new_leaf(r)
    ttm = new_leaf(T)
    p, d, v = bs_analytic(spot, strike, vol, rate, ttm)
    values.append([p, d, v])
```

```python
df = pd.DataFrame(data=values, columns=["pv", "delta", "vega"])
df["spot"] = spots
```

```python
my_plot(df, "spot", "pv")
```

```python
my_plot(df, "spot", "delta")
```

```python
my_plot(df, "spot", "vega")
```

```python
df.to_csv("analytic.csv", index=False)
```
