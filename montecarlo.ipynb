```python
import numpy as np
import pandas as pd
from main import my_plot, bs_montecarlo
```

```python
values = np.linspace(40.00, 220.00, 32)
paths = 100000  # to match greeks properly
rate = 0.02
T = 1.00
disc = np.exp(-T * rate)
v0 = 0.40
K = 100.00

a = np.zeros(shape=(len(values), 4))
for i, S in enumerate(values):
    pv, delta, vega = bs_montecarlo(S, K, v0, num_paths=paths, rate=rate, T=T)
    a[i] = [disc * pv, disc * delta, disc * vega, S]
```

```python
df = pd.DataFrame(data=a, columns=["pv", "delta", "vega", "spot"])

my_plot(df, "spot", "pv")
```

```python
my_plot(df, "spot", "delta")
```

```python
my_plot(df, "spot", "vega")
```
